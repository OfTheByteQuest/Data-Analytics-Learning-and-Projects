### Plotting with Pandas

```python
fb = pd.read_csv(
    filepath_or_buffer = "/content/Data-Analytics-Learning-and-Projects/Chapter-5/Data/fb_stock_prices_2018.csv",
    index_col = 'date',
    parse_dates = True
)
quakes = pd.read_csv('/content/Data-Analytics-Learning-and-Projects/Chapter-5/Data/earthquakes.csv')
```

#### Evolution over time

```python
fb.plot(
    kind='line',
    y='open',
    label='Opening Price',
    figsize=(7.7, 3.75),
    grid=True,
    legend=True,
    title='Evolution of Facebook Open Price',
    style='b-'
)
```

```python
fb.iloc[:5,].plot(
    y=['open', 'high', 'low', 'close'],
    style=['b-o', 'r--', 'k:', 'g-.'],
    label = ['Opening Price', 'Highest Price', 'Lowest Price', 'Closing Price'],
    grid=True,
    title='Facebook OHLC Prices during 1st Week of Trading 2018',
    figsize=(7.7, 3.5)
)
```

```python
fb.plot(
    kind='line',
    subplots=True,
    layout=(3,2),
    figsize=(15,10),
    title='Facebook Stock 2018'
)
```

#### Relationship between variables

```python
fb.assign(
    max_abs_change = fb.high - fb.low
).plot(
    kind='scatter',
    x='volume',
    y = 'max_abs_change',
    title='Facebook Daily High - Low vs. Volume Traded'
)
```

```python
fb.assign(
    max_abs_change = fb.high - fb.low
).plot(
    kind='scatter',
    x='volume',
    y = 'max_abs_change',
    title='Facebook Daily High - Low vs. Volume Traded',
    logx=True
)
```

***Note:*** The `plot()` method from pandas has three arguments for log
scales: `logx/logy` for single axis adjustments and `loglog` for setting
both to the log scale

```python
fb.assign(
    max_abs_change = fb.high - fb.low
).plot(
    kind='scatter',
    x='volume',
    y = 'max_abs_change',
    title='Facebook Daily High - Low vs. Volume Traded',
    logx=True,
    alpha=0.25,
)
```

```python
fb.assign(
    volume=np.log(fb.volume), # Pre-transformation is required for hexbin plot
    max_abs_change = fb.high - fb.low
).plot(
    kind='hexbin',
    x='volume',
    y = 'max_abs_change',
    title='Facebook Daily High - Low vs. log(Volume Traded)',
    colormap='gray_r',
    gridsize=20,
    # logx=True, ## Somehow this doesn't work for hexbin plot type
    sharex=False
)
```
