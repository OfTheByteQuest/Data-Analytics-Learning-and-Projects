```python
fb = pd.read_csv(
    filepath_or_buffer="/content/Data-Analytics-Learning-and-Projects/Chapter-5/Data/fb_stock_prices_2018.csv",
    index_col='date',
    parse_dates=True
)
```

```python
quakes = pd.read_csv('/content/Data-Analytics-Learning-and-Projects/Chapter-5/Data/earthquakes.csv')
plt.hist(quakes.query('magType == "ml"').mag, bins='auto')
```

```python
x = quakes.query('magType == "ml"').mag
fig, ax = plt.subplots(1,2,figsize=(10,3))
for ax, bins in zip(ax, [7, 35]):
  ax.hist(x,bins=bins)
  ax.set_title(f'Bins param: {bins}')
```

#### Plot Components

```python
fig = plt.figure()
```

```python
fig, axes = plt.subplots(1,2)
```

```python
fig = plt.figure(figsize=(3,3))
outside = fig.add_axes([0.1,0.1,0.9,0.9])
inside = fig.add_axes([0.7,0.7, 0.25, 0.25])
```

```python
fig = plt.figure(figsize=(8,8))
gs = fig.add_gridspec(3,3)
top_left = fig.add_subplot(gs[0,0])
mid_left = fig.add_subplot(gs[1,0])
top_right = fig.add_subplot(gs[:2, 1:])
bottom = fig.add_subplot(gs[2,:])
```
