##### **Behavior of Estimators: Evaluating Values Outside the Sample in Non-Parametric Estimators**

**1. Overview**
Non-parametric estimators like ECDF (Empirical Cumulative Distribution Function) and KDE (Kernel Density Estimation) are statistical tools that estimate distribution properties without assuming any specific parametric form. A key feature of these estimators is their ability to evaluate values even outside the range of the observed data.

###### **2. ECDF (Empirical Cumulative Distribution Function)**

**Definition:**
- The ECDF at a point `x` is defined as the proportion of data points in the sample that are less than or equal to `x`.

**How it works:**
- It constructs a **step function** based on the sorted sample.
- For any value of `x`, the ECDF counts how many data points satisfy `data ≤ x`, divided by the total number of points `n`.

**Behavior outside sample data:**
- Even if `x` is not in the sample, the ECDF can still evaluate the proportion of data less than or equal to it.
- For `x` less than the minimum data point, ECDF returns `0`.
- For `x` greater than the maximum, ECDF returns `1`.
- For intermediate values, it returns a step-wise increase.

**Key Point:**
- **No smoothing** is applied, and the output is a piecewise constant function.

###### **3. KDE (Kernel Density Estimation)**

**Definition:**
- KDE estimates the **probability density function** (PDF) by placing a smooth kernel (e.g., Gaussian) at each data point and summing these kernels.

**How it works:**
- Each data point becomes a smooth “bump” (kernel).
- These bumps overlap and create a smooth curve when added together.
- The bandwidth (smoothing parameter) controls the width of the bumps.

**Behavior outside sample data:**
- Since kernels like the Gaussian have infinite support, they contribute to values even far from their center.
- This allows KDE to give non-zero values for points outside the original data range.
- The density gradually decays away from the data but remains defined for all real numbers.

**Key Point:**
- KDE is **smoothed**, unlike ECDF, and yields a **continuous density estimate**.

###### **4. Comparison Table**

| Estimator | Smoothed | Returns values outside data? | Output Type           | Behavior Beyond Range       |
|-----------|----------|------------------------------|------------------------|-----------------------------|
| ECDF      | No       | Yes                          | Step function (0 to 1) | Constant at 0 or 1 outside  |
| KDE       | Yes      | Yes                          | Smooth density curve   | Tails decay gradually       |

```python
# Recreated in the following code-block

from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(quakes.query('magType == "ml"').mag)
plt.plot(ecdf.x, ecdf.y)
plt.xlabel('Magnitude Of ml type Earthquakes')
plt.ylabel('Cumulative Probability')
plt.title("ECDF of earthquake magnitude with magType ml")
```

```python
from scipy.stats import ecdf

res = ecdf(quakes.query('magType == "ml"').mag)
ax = plt.subplot()
res.cdf.plot(ax)
ax.set_xlabel('Magnitude Of ml type Earthquakes')
ax.set_ylabel('Cumulative Probability')
ax.set_title("ECDF of earthquake magnitude with magType ml")
```

***Note:***Remember, slicing notation is of the form `[start:stop:step]`,

```python
quakes.parsed_place.value_counts().iloc[14::-1].plot(
    kind='barh',
    figsize=(10,5),
    title='Top 15 Places for Earthquakes'\
          '(September 18, 2018 -October 13, 2018)',
    ylabel=''
)
plt.xlabel('Earthquakes')
```

```python
quakes.groupby(
    'parsed_place'
).tsunami.sum().sort_values().iloc[-10::,].plot( # Since, the sorting by default happens in ascending order so, we takes values from bottom
    kind='barh',
    figsize=(10,5),
    title='Top 10 Places for Tsunamis'\
          '(September 18, 2018 -October 13, 2018)',
    ylabel=''
)
plt.xlabel('Tsunamis')
```

```python
indonesia_quakes = quakes.query(
    'parsed_place == "Indonesia"'
).assign(
    time = lambda x: pd.to_datetime(x.time, unit='ms'),
    earthquake = 1
).set_index('time').resample('1D').sum()

indonesia_quakes.index = indonesia_quakes.index.strftime('%b\n%d') # Converts date time into month_name\nday, which allows to enter escape newline character in tick names

indonesia_quakes.plot(
    kind='bar',
    y = ['earthquake', 'tsunami'],
    figsize=(15,3),
    rot=0,
    label=['Earthquakes', 'Tsunamis'],
    title='Eathquakes and Tsunamis in Indonesia'\
          '(September 18, 2018 - October 13, 2018)'
)
plt.xlabel('Date')
plt.ylabel('Count')
```

```python
quakes.magType.value_counts().plot(
kind='bar', title='Earthquakes Recorded per magType', rot=0
)
plt.xlabel('magType')
plt.ylabel('earthquakes')
```

```python
quakes[
    quakes.parsed_place.isin(
        ['California', 'Alaska', 'Nevada', 'Hawaii']
    )].groupby(
        ['parsed_place', 'magType']
    ).mag.count().unstack().plot(
        kind='bar',
        column=['mb', 'md', 'mh', 'ml','mw'],
        title='EmagTypes used in top 4 places with earthquakes',
        )
plt.xlabel('magType')
plt.ylabel('earthquakes')
```

```python
pivot = quakes.assign(
    mag_bin = lambda x: np.floor(x.mag)
).pivot_table(
    index = 'mag_bin',
    columns = 'magType',
    values = 'mag',
    aggfunc = 'count'
)

pivot.plot.bar(
    stacked=True,
    rot=0,
    title = 'Earthquakes by integer magnitude and magType'
)
plt.ylabel('earthquakes')
```

```python
normalized_pivot = pivot.fillna(0).apply(
    lambda x: x/x.sum(), axis=1
    )
ax = normalized_pivot.plot.bar(
    stacked=True, rot=0, figsize=(10, 5),
    title='Percentage of earthquakes by integer magnitude '\
          'for each magType'
    )

ax.legend(bbox_to_anchor=(1, 0.8)) # move legend to right of plot
plt.ylabel('percentage')
```

#### Lag plots

```python
from pandas.plotting import lag_plot
rng = np.random.default_rng(seed=0)
lag_plot(pd.Series(rng.uniform(size=200)))
```

```python
# In the below code-block this result is recreated

lag_plot(fb.close, s=15, alpha=0.4, edgecolor='k')
```

```python
fb.assign(
    close_lag = lambda x: x.close.shift()
).plot.scatter(
    x='close_lag',
    y='close',
    title='Lag Plot of Facebook Closing Prices',
    alpha=0.4,
    edgecolor='k'
)
```

***Explanation:*** In the below code we have tried to map the closing price of Fb's stocks on a specific day with the closing price of that day previosu week. Since, the market is open for only 5 days a week we are using the lag of 5 days.

```python
lag_plot(fb.close, lag=5, s=15, alpha=0.4, edgecolor='k')
```
